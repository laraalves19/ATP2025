a) Contar ocorrências: defina uma função que recebe uma lista e um elemento como argumento e devolve
o número de ocorrências desse elemento na lista.

```python
def contarocor(lista,n):
    res=0
    for elem in lista:
        if elem==n:
            res=res+1
    return res

contarocor([1,2,3,2],2)
```

b) Comum?: defina uma função que recebe duas listas como argumento e devolve o valor booleano
verdadeiro se as listas tiverem pelo menos um elemento comum e falso caso contrário.

```python
def comum(lista1,lista2):
    cond=False
    for elem in lista1:
        while cond==False:
            if elem in lista2:
                cond=True
            else:
                cond=False 
    return cond

comum([1,2,3,4],[1,5,6])
```

a) splitLista: defina uma função que recebe dois argumentos, uma lista de números e um número, e dá
como resultado um tuplo composto por duas listas, a primeira contendo os números menores ou iguais ao
número passado como argumento, e a segunda contendo os números maiores que o argumento.

```python
def splitLista(lista,n):
    lista_menores=[]
    lista_maiores=[]
    for elem in lista:
        if elem<=n:
            lista_menores.append(elem)
        elif elem>n:
            lista_maiores.append(elem)
    res=(lista_menores, lista_maiores)
    return res 

splitLista([1,2,3,4,5], 3)

```

```python
def multiplos23(n):
    res=0
    i=1
    while i>=1 and i<=n:
        if 23%i==0:
            res=res+1
            i=i+1
        else:
            i=i+1
    return res 

print(multiplos23(100))

        
```

c) max3prod: defina uma função que recebe uma lista de números inteiros positivos e devolve o maior
produto que for possível calcular multiplicando os 3 maiores inteiros da lista.

```python
def max3prod(lista):
    res=1
    lista.sort()
    for elem in lista:
        res= lista[-3]*lista[-2]*lista[-1]
    return res

print(max3prod([5,3,2,4,1]))

        
```

d) reduxInt: defina uma função que recebe um número inteiro positivo como argumento e vai adicionar
repetidamente os seu dígitos até obter apenas um dígito que é retornado como resultado.

```python
def redeuxInf(n):
    soma=0
    n_str=str(n)
    for c in n_str:
        num=int(c)
        soma=soma+num
    return soma

print(reduxInt(998))
```
